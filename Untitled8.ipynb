```python
# ==========================================
# [1] الإعدادات وربط Google Drive
# ==========================================
import os, cv2, random, shutil
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from glob import glob
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from scipy import stats
import matplotlib.pyplot as plt
from skimage.filters import frangi, sato, sobel
import warnings
warnings.filterwarnings("ignore")

# ربط جوجل درايف
from google.colab import drive
drive.mount('/content/drive')

# الإعدادات
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ الجهاز المستخدم: {device}")

# المسارات - عدل DRIVE_ROOT للمسار الصحيح في درايفك
DRIVE_ROOT = "/content/drive/MyDrive/FIVES"
SAVE_ROOT = "/content/FIVES_Processed_512" # الحفظ في مساحة كولاب السريعة

TRAIN_IMG_DIR = os.path.join(SAVE_ROOT, "train", "images")
TRAIN_MASK_DIR = os.path.join(SAVE_ROOT, "train", "masks")
VESSEL_DIR_TRAIN = os.path.join(SAVE_ROOT, "train", "vessels")

TEST_IMG_DIR = os.path.join(SAVE_ROOT, "test", "images")
TEST_MASK_DIR = os.path.join(SAVE_ROOT, "test", "masks")
VESSEL_DIR_TEST = os.path.join(SAVE_ROOT, "test", "vessels")

IMG_SIZE = 512
VESSEL_CHANNELS = 1
USE_MULTI_CHANNEL = (VESSEL_CHANNELS == 3)
```

```python
# ==========================================
# [2] تجهيز البيانات والحساب المسبق (Data Preprocessing & Precompute)
# ==========================================
if os.path.exists(SAVE_ROOT):
    shutil.rmtree(SAVE_ROOT)
    print("🗑️ تم مسح المجلد القديم.")

for d in [TRAIN_IMG_DIR, TRAIN_MASK_DIR, VESSEL_DIR_TRAIN, TEST_IMG_DIR, TEST_MASK_DIR, VESSEL_DIR_TEST]:
    os.makedirs(d, exist_ok=True)

def process_and_save(img_path, mask_path, save_img_dir, save_mask_dir):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    if img is None or mask is None: return

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img = clahe.apply(img)

    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_LINEAR)
    mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_NEAREST)
    _, mask = cv2.threshold(mask, 127, 255, cv2.THRESH_BINARY)

    cv2.imwrite(os.path.join(save_img_dir, os.path.basename(img_path)), img)
    cv2.imwrite(os.path.join(save_mask_dir, os.path.basename(mask_path)), mask)

def precompute_vessels(img_dir, vessel_dir, use_multi_channel):
    for img_name in tqdm(os.listdir(img_dir), desc=f"Precomputing Vessels in {os.path.basename(os.path.dirname(img_dir))}"):
        img_path = os.path.join(img_dir, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0

        f_map = frangi(img, sigmas=range(1, 6), black_ridges=False, beta=0.5, gamma=15)
        f_map = (f_map - f_map.min()) / (f_map.max() - f_map.min() + 1e-8)

        if use_multi_channel:
            s_map = sato(img, sigmas=range(1, 6), black_ridges=False)
            s_map = (s_map - s_map.min()) / (s_map.max() - s_map.min() + 1e-8)
            g_map = sobel(img)
            g_map = (g_map - g_map.min()) / (g_map.max() - g_map.min() + 1e-8)
            v_map = np.stack([f_map, s_map, g_map], axis=0)
        else:
            v_map = np.expand_dims(f_map, axis=0)

        npy_name = img_name.replace('.png', '.npy')
        np.save(os.path.join(vessel_dir, npy_name), v_map)

print(f"⏳ جاري تجهيز الصور بدقة {IMG_SIZE}x{IMG_SIZE}...")
train_images = glob(os.path.join(ROOT, "train", "Original", "*.png"))
for img_p in tqdm(train_images, desc="Training Set"):
    process_and_save(img_p, img_p.replace("Original", "Ground truth"), TRAIN_IMG_DIR, TRAIN_MASK_DIR)

test_images = glob(os.path.join(ROOT, "test", "Original", "*.png"))
for img_p in tqdm(test_images, desc="Test Set"):
    process_and_save(img_p, img_p.replace("Original", "Ground truth"), TEST_IMG_DIR, TEST_MASK_DIR)

print("\n⏳ جاري الحساب المسبق لخرائط الأوعية (Precomputing Vessel Maps)...")
precompute_vessels(TRAIN_IMG_DIR, VESSEL_DIR_TRAIN, USE_MULTI_CHANNEL)
precompute_vessels(TEST_IMG_DIR, VESSEL_DIR_TEST, USE_MULTI_CHANNEL)
```

```python

# ==========================================
# [3] بناء الـ Dataset (Vessel Maps + DataLoader)
# ==========================================
class FastDataset(Dataset):
    def __init__(self, file_names, split='train', use_multi_channel=False):
        self.file_names = file_names
        img_dir = TRAIN_IMG_DIR if split == 'train' else TEST_IMG_DIR
        mask_dir = TRAIN_MASK_DIR if split == 'train' else TEST_MASK_DIR
        vessel_dir = VESSEL_DIR_TRAIN if split == 'train' else VESSEL_DIR_TEST

        self.images, self.vessel_maps, self.masks = [], [], []
        print(f"⏳ Loading {split} data to RAM...")

        for fn in tqdm(self.file_names):
            img = cv2.imread(os.path.join(img_dir, fn), cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0
            mask = cv2.imread(os.path.join(mask_dir, fn), cv2.IMREAD_GRAYSCALE)
            mask = (mask > 127).astype(np.float32)

            npy_name = fn.replace('.png', '.npy')
            v_map = np.load(os.path.join(vessel_dir, npy_name))

            self.images.append(np.expand_dims(img, axis=0))
            self.vessel_maps.append(v_map)
            self.masks.append(np.expand_dims(mask, axis=0))

    def __len__(self): return len(self.file_names)
    def __getitem__(self, idx):
        return (
            torch.FloatTensor(self.images[idx]),
            torch.FloatTensor(self.vessel_maps[idx]),
            torch.FloatTensor(self.masks[idx])
        )

def get_paired_files(img_dir, mask_dir):
    return sorted(list(set(os.listdir(img_dir)) & set(os.listdir(mask_dir))))

train_files = get_paired_files(TRAIN_IMG_DIR, TRAIN_MASK_DIR)
if len(train_files) == 0: raise ValueError("⚠️ مسار ROOT خاطئ، لم يتم العثور على صور!")
train_split, val_split = train_test_split(train_files, test_size=0.1, random_state=SEED)
test_files = get_paired_files(TEST_IMG_DIR, TEST_MASK_DIR)

train_loader = DataLoader(FastDataset(train_split, 'train', USE_MULTI_CHANNEL), batch_size=4, shuffle=True)
val_loader   = DataLoader(FastDataset(val_split, 'train', USE_MULTI_CHANNEL), batch_size=4, shuffle=False)
test_loader  = DataLoader(FastDataset(test_files, 'test', USE_MULTI_CHANNEL), batch_size=4, shuffle=False)
```

```python

# ==========================================
# [4] المعماريات ودوال الخسارة (Architectures & Losses)
# ==========================================
class DoubleConvV2(nn.Module):
    def __init__(self, in_c, out_c):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_c, out_c, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_c), nn.ReLU(inplace=True),
            nn.Dropout2d(0.1),
            nn.Conv2d(out_c, out_c, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_c), nn.ReLU(inplace=True)
        )
    def forward(self, x): return self.conv(x)

class BaselineUNetV2(nn.Module):
    def __init__(self):
        super().__init__()
        self.inc = DoubleConvV2(1, 64)
        self.down1 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(64, 128))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(128, 256))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(256, 512))

        self.up1 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.conv1 = DoubleConvV2(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.conv2 = DoubleConvV2(256, 128)
        self.up3 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.conv3 = DoubleConvV2(128, 64)
        self.outc = nn.Conv2d(64, 1, 1)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        d = self.conv1(torch.cat([self.up1(x4), x3], dim=1))
        d = self.conv2(torch.cat([self.up2(d), x2], dim=1))
        d = self.conv3(torch.cat([self.up3(d), x1], dim=1))
        return torch.sigmoid(self.outc(d))

class SpatialChannelAttention(nn.Module):
    def __init__(self, in_channels, vessel_channels):
        super().__init__()
        self.spatial = nn.Sequential(
            nn.Conv2d(in_channels + vessel_channels, in_channels, 3, padding=1),
            nn.BatchNorm2d(in_channels), nn.ReLU(inplace=True),
            nn.Conv2d(in_channels, in_channels, 1), nn.Sigmoid()
        )
        self.channel = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_channels, in_channels // 4, 1), nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, in_channels, 1), nn.Sigmoid()
        )
        self.alpha = nn.Parameter(torch.tensor(0.0))

    def forward(self, x, v):
        v_res = F.interpolate(v, size=x.shape[2:], mode='bilinear', align_corners=False)
        s_att = self.spatial(torch.cat([x, v_res], dim=1))
        c_att = self.channel(x)
        w = torch.sigmoid(self.alpha)
        return (1.0 - w) * x + w * (x * s_att * c_att)

class DSBlock(nn.Module):
    def __init__(self, in_c):
        super().__init__()
        self.ds = nn.Sequential(
            nn.Conv2d(in_c, 64, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 1, 1)
        )
    def forward(self, x): return self.ds(x)

class PCGNetV2(nn.Module):
    def __init__(self, v_channels=VESSEL_CHANNELS):
        super().__init__()
        self.inc = DoubleConvV2(1, 64)
        self.down1 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(64, 128))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(128, 256))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(256, 512))

        self.bottleneck_att = SpatialChannelAttention(512, v_channels)
        self.att1 = SpatialChannelAttention(256, v_channels)
        self.att2 = SpatialChannelAttention(128, v_channels)
        self.att3 = SpatialChannelAttention(64, v_channels)

        self.up1 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.conv1 = DoubleConvV2(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.conv2 = DoubleConvV2(256, 128)
        self.up3 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.conv3 = DoubleConvV2(128, 64)
        self.outc = nn.Conv2d(64, 1, 1)

        self.ds1 = DSBlock(256)
        self.ds2 = DSBlock(128)

    def forward(self, image, v_map):
        x1 = self.inc(image)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)

        x4 = self.bottleneck_att(x4, v_map)

        d = self.up1(x4)
        d = self.conv1(torch.cat([d, self.att1(x3, v_map)], dim=1))
        ds1_out = self.ds1(d)

        d = self.up2(d)
        d = self.conv2(torch.cat([d, self.att2(x2, v_map)], dim=1))
        ds2_out = self.ds2(d)

        d = self.up3(d)
        d = self.conv3(torch.cat([d, self.att3(x1, v_map)], dim=1))
        main_out = self.outc(d)

        if self.training:
            return torch.sigmoid(main_out), torch.sigmoid(ds1_out), torch.sigmoid(ds2_out)
        return torch.sigmoid(main_out)

    def get_alphas(self):
        return {
            'bottleneck': torch.sigmoid(self.bottleneck_att.alpha).item(),
            'deep': torch.sigmoid(self.att1.alpha).item(),
            'mid':  torch.sigmoid(self.att2.alpha).item(),
            'fine': torch.sigmoid(self.att3.alpha).item(),
        }

class SoftSkeletonize(nn.Module):
    def __init__(self, num_iter=10):
        super().__init__()
        self.num_iter = num_iter
    def soft_erode(self, img): return -F.max_pool2d(-img, 3, stride=1, padding=1)
    def soft_dilate(self, img): return F.max_pool2d(img, 3, stride=1, padding=1)
    def soft_open(self, img): return self.soft_dilate(self.soft_erode(img))
    def forward(self, img):
        img_open = self.soft_open(img)
        skel = F.relu(img - img_open)
        curr_img = img
        for _ in range(self.num_iter):
            curr_img = self.soft_erode(curr_img)
            curr_open = self.soft_open(curr_img)
            delta = F.relu(curr_img - curr_open)
            skel = skel + F.relu(delta - skel * delta)
        return skel

class TrueSoftCLDiceLoss(nn.Module):
    def __init__(self, smooth=1.):
        super().__init__()
        self.smooth = smooth
        self.skel = SoftSkeletonize(num_iter=10)
    def forward(self, y_pred, y_true):
        skel_true = self.skel(y_true)
        skel_pred = self.skel(y_pred)
        tprec = (torch.sum(skel_pred * y_true) + self.smooth) / (torch.sum(skel_pred) + self.smooth)
        tsens = (torch.sum(skel_true * y_pred) + self.smooth) / (torch.sum(skel_true) + self.smooth)
        cl_dice = 2.0 * (tprec * tsens) / (tprec + tsens)
        return 1.0 - cl_dice

class FocalLoss(nn.Module):
    def __init__(self, alpha=0.75, gamma=2.0):
        super().__init__()
        self.alpha, self.gamma = alpha, gamma
    def forward(self, inputs, targets):
        bce = F.binary_cross_entropy(inputs, targets, reduction='none')
        p_t = inputs * targets + (1 - inputs) * (1 - targets)
        alpha_t = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        loss = alpha_t * (1 - p_t) ** self.gamma * bce
        return loss.mean()

class PCG_AdvancedLossV2(nn.Module):
    def __init__(self, cl_weight=0.3):
        super().__init__()
        self.cl_weight = cl_weight
        self.focal = FocalLoss()
        self.cldice = TrueSoftCLDiceLoss()
    def forward(self, pred, target):
        focal_loss = self.focal(pred, target)
        dice_loss = 1 - (2*(pred*target).sum() + 1e-8) / (pred.sum() + target.sum() + 1e-8)
        cld = self.cldice(pred, target)
        return (1 - self.cl_weight) * (focal_loss + dice_loss) + self.cl_weight * cld
```

```python
# ==========================================
# [5] التدريب المتقدم (Advanced Training Loop)
# ==========================================
def train_model(model, train_loader, val_loader, device, model_name, epochs=80, patience=20):
    optimizer = torch.optim.AdamW(model.parameters(), lr=5e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer, max_lr=5e-4, epochs=epochs, steps_per_epoch=len(train_loader),
        pct_start=0.1, anneal_strategy='cos'
    )

    criterion = PCG_AdvancedLossV2(cl_weight=0.3)
    best_dice, counter, is_pcg = 0.0, 0, "PCG" in model_name
    save_path = f"{model_name}_BEST.pth"
    history = {'train_loss': [], 'val_dice': []}

    print(f"\n🚀 Training {model_name} | AdamW + OneCycleLR | Deep Sup: {'ON' if is_pcg else 'OFF'}")

    for epoch in range(epochs):
        model.train(); epoch_loss = 0
        ds_weight = max(0.1, 0.4 * (1 - epoch/epochs))

        for imgs, v_maps, masks in train_loader:
            imgs, v_maps, masks = imgs.to(device), v_maps.to(device), masks.to(device)
            optimizer.zero_grad()

            if is_pcg:
                main_out, ds1_out, ds2_out = model(imgs, v_maps)
                l_main = criterion(main_out, masks)
                l_ds1 = criterion(ds1_out, F.interpolate(masks, size=ds1_out.shape[2:], mode='nearest'))
                l_ds2 = criterion(ds2_out, F.interpolate(masks, size=ds2_out.shape[2:], mode='nearest'))
                loss = l_main + ds_weight * (l_ds1 + l_ds2)
            else:
                loss = criterion(model(imgs), masks)

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()
            epoch_loss += loss.item()

        avg_loss = epoch_loss / len(train_loader)
        history['train_loss'].append(avg_loss)

        # Validation
        model.eval(); val_dices = []
        with torch.no_grad():
            for imgs, v_maps, masks in val_loader:
                imgs, v_maps, masks = imgs.to(device), v_maps.to(device), masks.to(device)
                out = model(imgs, v_maps) if is_pcg else model(imgs)

                p = (out.squeeze(1).cpu().numpy() > 0.5).astype(bool)
                t = masks.squeeze(1).cpu().numpy().astype(bool)

                for j in range(p.shape[0]):
                    tp = np.logical_and(p[j], t[j]).sum()
                    fp = np.logical_and(p[j], ~t[j]).sum()
                    fn = np.logical_and(~p[j], t[j]).sum()
                    val_dices.append((2*tp) / (2*tp + fp + fn + 1e-8))

        avg_dice = np.mean(val_dices)
        history['val_dice'].append(avg_dice)

        marker = " ⭐ BEST" if avg_dice > best_dice else ""
        if marker: best_dice, counter = avg_dice, 0; torch.save(model.state_dict(), save_path)
        else: counter += 1

        print(f"  Ep [{epoch+1:2d}/{epochs}] Loss: {avg_loss:.4f} | Val Dice: {avg_dice:.4f} | LR: {optimizer.param_groups[0]['lr']:.6e} | DS_W: {ds_weight:.2f}{marker}")
        if counter >= patience: print("  ⏹️ Early Stop"); break

    return save_path, history

b_model = BaselineUNetV2().to(device)
path_b, h_base = train_model(b_model, train_loader, val_loader, device, "Baseline_UNetV2")

p_model = PCGNetV2().to(device)
path_p, h_pcg = train_model(p_model, train_loader, val_loader, device, "PCGNetV2")

# رسم منحنيات التدريب
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.plot(h_base['train_loss'], label='Baseline', linewidth=2)
ax1.plot(h_pcg['train_loss'], label='PCG-Net', linewidth=2)
ax1.set_title('Training Loss', fontsize=14, fontweight='bold')
ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss'); ax1.legend(); ax1.grid(True)

ax2.plot(h_base['val_dice'], label='Baseline', linewidth=2)
ax2.plot(h_pcg['val_dice'], label='PCG-Net', linewidth=2)
ax2.set_title('Validation Dice', fontsize=14, fontweight='bold')
ax2.set_xlabel('Epoch'); ax2.set_ylabel('Dice'); ax2.legend(); ax2.grid(True)
plt.tight_layout(); plt.show()
```

```python
# ==========================================
# [6] التقييم العادل والمقارنة (Fair Evaluation)
# ==========================================
b_model.load_state_dict(torch.load('Baseline_UNetV2_BEST.pth', map_location=device))
b_model.eval()
p_model.load_state_dict(torch.load('PCGNetV2_BEST.pth', map_location=device))
p_model.eval()

def get_error_map(pred, target):
    err = np.zeros((pred.shape[0], pred.shape[1], 3), dtype=np.uint8)
    err[(pred==1) & (target==1)] = [255, 255, 255]
    err[(pred==1) & (target==0)] = [255, 0, 0]
    err[(pred==0) & (target==1)] = [0, 255, 255]
    return err

def remove_small_noise(mask_bool, min_size=50):
    mask_u8 = (mask_bool * 255).astype(np.uint8)
    num, labels, stats, _ = cv2.connectedComponentsWithStats(mask_u8, connectivity=8)
    clean = np.zeros_like(mask_u8)
    for i in range(1, num):
        if stats[i, cv2.CC_STAT_AREA] >= min_size: clean[labels == i] = 1
    return clean.astype(bool)

def visualize_fair(loader, b_mod, p_mod, dev, num=4):
    fig, axes = plt.subplots(num, 6, figsize=(24, 4 * num))
    fig.suptitle("Fair Comparison: Threshold=0.5 | Post-process: Remove areas < 50px", fontsize=16, fontweight='bold')
    cols = ['Original', 'Ground Truth', 'Baseline Pred', 'Baseline Error', 'PCG-Net Pred', 'PCG-Net Error']
    for ax, c in zip(axes[0], cols): ax.set_title(c, fontweight='bold')

    with torch.no_grad():
        for i, (imgs, vmaps, masks) in enumerate(loader):
            if i >= num: break
            img, vmap, mask = imgs[0:1].to(dev), vmaps[0:1].to(dev), masks[0:1].to(dev)

            p_base_raw = b_mod(img).squeeze(0).squeeze(0).cpu().numpy() > 0.5
            p_pcg_raw  = p_mod(img, vmap).squeeze(0).squeeze(0).cpu().numpy() > 0.5

            p_b_clean = remove_small_noise(p_base_raw, 50)
            p_p_clean = remove_small_noise(p_pcg_raw, 50)

            t_mask = mask.squeeze(0).squeeze(0).cpu().numpy().astype(bool)
            img_vis = img.squeeze(0).squeeze(0).cpu().numpy()

            err_b = get_error_map(p_b_clean, t_mask)
            err_p = get_error_map(p_p_clean, t_mask)

            axes[i,0].imshow(img_vis, cmap='gray'); axes[i,0].axis('off')
            axes[i,1].imshow(t_mask, cmap='gray'); axes[i,1].axis('off')
            axes[i,2].imshow(p_b_clean, cmap='gray'); axes[i,2].axis('off')
            axes[i,3].imshow(err_b); axes[i,3].axis('off')
            axes[i,4].imshow(p_p_clean, cmap='gray'); axes[i,4].axis('off')
            axes[i,5].imshow(err_p); axes[i,5].axis('off')
    plt.tight_layout()
    plt.show()

print("\n🖼️ جاري استخراج الصور العادلة...")
visualize_fair(test_loader, b_model, p_model, device, num=4)
```

```python
# ==========================================
# [7] النتائج النهائية والإحصائيات (Final Stats)
# ==========================================
def compute_metrics(pred, true):
    tp = np.logical_and(pred, true).sum()
    fp = np.logical_and(pred, ~true).sum()
    fn = np.logical_and(~pred, true).sum()
    tn = np.logical_and(~pred, ~true).sum()
    return {
        'Dice': (2*tp) / (2*tp + fp + fn + 1e-8),
        'IoU': tp / (tp + fp + fn + 1e-8),
        'Sensitivity': tp / (tp + fn + 1e-8),
        'Specificity': tn / (tn + fp + 1e-8),
        'Precision': tp / (tp + fp + 1e-8),
        'Accuracy': (tp + tn) / (tp + tn + fp + fn + 1e-8)
    }

print("\n⏳ جاري حساب الإحصائيات لجميع صور الاختبار...")

base_metrics = {'Dice': [], 'IoU': [], 'Sensitivity': [], 'Specificity': [], 'Precision': [], 'Accuracy': []}
pcg_metrics = {'Dice': [], 'IoU': [], 'Sensitivity': [], 'Specificity': [], 'Precision': [], 'Accuracy': []}

with torch.no_grad():
    for imgs, vmaps, masks in tqdm(test_loader):
        imgs, vmaps, masks = imgs.to(device), vmaps.to(device), masks.to(device)

        pb_raw = b_model(imgs).squeeze(1).cpu().numpy() > 0.5
        pp_raw = p_model(imgs, vmaps).squeeze(1).cpu().numpy() > 0.5
        t_masks = masks.squeeze(1).cpu().numpy().astype(bool)

        for i in range(t_masks.shape[0]):
            t = t_masks[i]

            pb = remove_small_noise(pb_raw[i], 50)
            bm = compute_metrics(pb, t)
            for k in base_metrics: base_metrics[k].append(bm[k])

            pp = remove_small_noise(pp_raw[i], 50)
            pm = compute_metrics(pp, t)
            for k in pcg_metrics: pcg_metrics[k].append(pm[k])

stat, p_val = stats.wilcoxon(base_metrics['Dice'], pcg_metrics['Dice'])

print("\n" + "="*65)
print(f"🔬 P-Value (Wilcoxon Signed-Rank Test on Dice): {p_val:.4e}")
if p_val < 0.05: print("✅ النتيجة: تفوق PCG-Net ذو دلالة إحصائية قوية!")
else: print("⚠️ النتيجة: لا توجد دلالة إحصائية قوية للتفوق.")
print("="*65)

alphas = p_model.get_alphas()
print("\n🧠 XAI Attention Weights (Sigmoid Alpha):")
for k, v in alphas.items(): print(f" - {k.capitalize()}: {v:.4f}")

plt.figure(figsize=(8, 5))
colors = ['#e74c3c', '#ff9999', '#66b3ff', '#99ff99']
bars = plt.bar(list(alphas.keys()), list(alphas.values()), color=colors, edgecolor='black')
plt.ylim(0, 1.0)
plt.title("XAI: Learned Attention Weights per Level", fontsize=14, fontweight='bold')
plt.ylabel('Sigmoid(α)', fontsize=12)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
             f'{bar.get_height():.4f}', ha='center', va='bottom', fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# [1] المكاتب والإعدادات (Imports & Setup)
# ==========================================
import os, cv2, random, shutil
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from glob import glob
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from scipy import stats
import matplotlib.pyplot as plt
from skimage.filters import frangi, sato, sobel
import warnings
warnings.filterwarnings("ignore")

# 🟢 ربط جوجل درايف
from google.colab import drive
drive.mount('/content/drive')

# التثبيت العشوائي
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ Using device: {device}")

# ==========================================
# 🟢 المسارات المخصصة لـ Google Colab
# ==========================================
ROOT = "/content/drive/MyDrive/FIVES A Fundus Image Dataset for AI-based Vessel Segmentation" # ⚠️ تأكد أن هذا هو اسم مجلد الداتا في درايفك
SAVE_ROOT = "/content/FIVES_Processed_512" # الحفظ في مساحة كولاب المؤقتة السريعة جداً

TRAIN_IMG_DIR = os.path.join(SAVE_ROOT, "train", "images")
TRAIN_MASK_DIR = os.path.join(SAVE_ROOT, "train", "masks")
VESSEL_DIR_TRAIN = os.path.join(SAVE_ROOT, "train", "vessels")

TEST_IMG_DIR = os.path.join(SAVE_ROOT, "test", "images")
TEST_MASK_DIR = os.path.join(SAVE_ROOT, "test", "masks")
VESSEL_DIR_TEST = os.path.join(SAVE_ROOT, "test", "vessels")

IMG_SIZE = 512
VESSEL_CHANNELS = 1 # غيّر لـ 3 لو أردت Frangi + Sato + Gradient
USE_MULTI_CHANNEL = (VESSEL_CHANNELS == 3)

# ==========================================
# [2] تجهيز البيانات والحساب المسبق (Data Preprocessing & Precompute)
# ==========================================
if os.path.exists(SAVE_ROOT):
    shutil.rmtree(SAVE_ROOT)
    print("🗑️ تم مسح المجلد القديم.")

for d in [TRAIN_IMG_DIR, TRAIN_MASK_DIR, VESSEL_DIR_TRAIN, TEST_IMG_DIR, TEST_MASK_DIR, VESSEL_DIR_TEST]:
    os.makedirs(d, exist_ok=True)

def process_and_save(img_path, mask_path, save_img_dir, save_mask_dir):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    if img is None or mask is None: return

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img = clahe.apply(img)

    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_LINEAR)
    mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_NEAREST)
    _, mask = cv2.threshold(mask, 127, 255, cv2.THRESH_BINARY)

    cv2.imwrite(os.path.join(save_img_dir, os.path.basename(img_path)), img)
    cv2.imwrite(os.path.join(save_mask_dir, os.path.basename(mask_path)), mask)

def precompute_vessels(img_dir, vessel_dir, use_multi_channel):
    for img_name in tqdm(os.listdir(img_dir), desc=f"Precomputing Vessels in {os.path.basename(os.path.dirname(img_dir))}"):
        img_path = os.path.join(img_dir, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0

        f_map = frangi(img, sigmas=range(1, 6), black_ridges=False, beta=0.5, gamma=15)
        f_map = (f_map - f_map.min()) / (f_map.max() - f_map.min() + 1e-8)

        if use_multi_channel:
            s_map = sato(img, sigmas=range(1, 6), black_ridges=False)
            s_map = (s_map - s_map.min()) / (s_map.max() - s_map.min() + 1e-8)
            g_map = sobel(img)
            g_map = (g_map - g_map.min()) / (g_map.max() - g_map.min() + 1e-8)
            v_map = np.stack([f_map, s_map, g_map], axis=0)
        else:
            v_map = np.expand_dims(f_map, axis=0)

        npy_name = img_name.replace('.png', '.npy')
        np.save(os.path.join(vessel_dir, npy_name), v_map)

print(f"⏳ جاري تجهيز الصور بدقة {IMG_SIZE}x{IMG_SIZE}...")
train_images = glob(os.path.join(ROOT, "train", "Original", "*.png"))
for img_p in tqdm(train_images, desc="Training Set"):
    process_and_save(img_p, img_p.replace("Original", "Ground truth"), TRAIN_IMG_DIR, TRAIN_MASK_DIR)

test_images = glob(os.path.join(ROOT, "test", "Original", "*.png"))
for img_p in tqdm(test_images, desc="Test Set"):
    process_and_save(img_p, img_p.replace("Original", "Ground truth"), TEST_IMG_DIR, TEST_MASK_DIR)

print("\n⏳ جاري الحساب المسبق لخرائط الأوعية (Precomputing Vessel Maps)...")
precompute_vessels(TRAIN_IMG_DIR, VESSEL_DIR_TRAIN, USE_MULTI_CHANNEL)
precompute_vessels(TEST_IMG_DIR, VESSEL_DIR_TEST, USE_MULTI_CHANNEL)

# ==========================================
# [3] بناء الـ Dataset (Vessel Maps + DataLoader)
# ==========================================
class FastDataset(Dataset):
    def __init__(self, file_names, split='train', use_multi_channel=False):
        self.file_names = file_names
        img_dir = TRAIN_IMG_DIR if split == 'train' else TEST_IMG_DIR
        mask_dir = TRAIN_MASK_DIR if split == 'train' else TEST_MASK_DIR
        vessel_dir = VESSEL_DIR_TRAIN if split == 'train' else VESSEL_DIR_TEST

        self.images, self.vessel_maps, self.masks = [], [], []
        print(f"⏳ Loading {split} data to RAM...")

        for fn in tqdm(self.file_names):
            img = cv2.imread(os.path.join(img_dir, fn), cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0
            mask = cv2.imread(os.path.join(mask_dir, fn), cv2.IMREAD_GRAYSCALE)
            mask = (mask > 127).astype(np.float32)

            npy_name = fn.replace('.png', '.npy')
            v_map = np.load(os.path.join(vessel_dir, npy_name))

            self.images.append(np.expand_dims(img, axis=0))
            self.vessel_maps.append(v_map)
            self.masks.append(np.expand_dims(mask, axis=0))

    def __len__(self): return len(self.file_names)
    def __getitem__(self, idx):
        # 🟢 التحويل إلى FloatTensor لحل مشكلة الـ GPU Float64 Crash
        return (
            torch.FloatTensor(self.images[idx]),
            torch.FloatTensor(self.vessel_maps[idx]),
            torch.FloatTensor(self.masks[idx])
        )

def get_paired_files(img_dir, mask_dir):
    return sorted(list(set(os.listdir(img_dir)) & set(os.listdir(mask_dir))))

train_files = get_paired_files(TRAIN_IMG_DIR, TRAIN_MASK_DIR)
if len(train_files) == 0: raise ValueError("⚠️ مسار ROOT خاطئ، لم يتم العثور على صور!")
train_split, val_split = train_test_split(train_files, test_size=0.1, random_state=SEED)
test_files = get_paired_files(TEST_IMG_DIR, TEST_MASK_DIR)

train_loader = DataLoader(FastDataset(train_split, 'train', USE_MULTI_CHANNEL), batch_size=4, shuffle=True)
val_loader   = DataLoader(FastDataset(val_split, 'train', USE_MULTI_CHANNEL), batch_size=4, shuffle=False)
test_loader  = DataLoader(FastDataset(test_files, 'test', USE_MULTI_CHANNEL), batch_size=4, shuffle=False)

# ==========================================
# [4] المعماريات ودوال الخسارة (Architectures & Losses)
# ==========================================
class DoubleConvV2(nn.Module):
    def __init__(self, in_c, out_c):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_c, out_c, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_c), nn.ReLU(inplace=True),
            nn.Dropout2d(0.1),
            nn.Conv2d(out_c, out_c, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_c), nn.ReLU(inplace=True)
        )
    def forward(self, x): return self.conv(x)

class BaselineUNetV2(nn.Module):
    def __init__(self):
        super().__init__()
        self.inc = DoubleConvV2(1, 64)
        self.down1 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(64, 128))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(128, 256))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(256, 512))

        self.up1 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.conv1 = DoubleConvV2(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.conv2 = DoubleConvV2(256, 128)
        self.up3 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.conv3 = DoubleConvV2(128, 64)
        self.outc = nn.Conv2d(64, 1, 1)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        d = self.conv1(torch.cat([self.up1(x4), x3], dim=1))
        d = self.conv2(torch.cat([self.up2(d), x2], dim=1))
        d = self.conv3(torch.cat([self.up3(d), x1], dim=1))
        return torch.sigmoid(self.outc(d))

class SpatialChannelAttention(nn.Module):
    def __init__(self, in_channels, vessel_channels):
        super().__init__()
        self.spatial = nn.Sequential(
            nn.Conv2d(in_channels + vessel_channels, in_channels, 3, padding=1),
            nn.BatchNorm2d(in_channels), nn.ReLU(inplace=True),
            nn.Conv2d(in_channels, in_channels, 1), nn.Sigmoid()
        )
        self.channel = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_channels, in_channels // 4, 1), nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, in_channels, 1), nn.Sigmoid()
        )
        self.alpha = nn.Parameter(torch.tensor(0.0))

    def forward(self, x, v):
        v_res = F.interpolate(v, size=x.shape[2:], mode='bilinear', align_corners=False)
        s_att = self.spatial(torch.cat([x, v_res], dim=1))
        c_att = self.channel(x)
        w = torch.sigmoid(self.alpha)
        return (1.0 - w) * x + w * (x * s_att * c_att)

class DSBlock(nn.Module):
    def __init__(self, in_c):
        super().__init__()
        self.ds = nn.Sequential(
            nn.Conv2d(in_c, 64, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 1, 1)
        )
    def forward(self, x): return self.ds(x)

class PCGNetV2(nn.Module):
    def __init__(self, v_channels=VESSEL_CHANNELS):
        super().__init__()
        self.inc = DoubleConvV2(1, 64)
        self.down1 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(64, 128))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(128, 256))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), DoubleConvV2(256, 512))

        self.bottleneck_att = SpatialChannelAttention(512, v_channels)
        self.att1 = SpatialChannelAttention(256, v_channels)
        self.att2 = SpatialChannelAttention(128, v_channels)
        self.att3 = SpatialChannelAttention(64, v_channels)

        self.up1 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.conv1 = DoubleConvV2(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.conv2 = DoubleConvV2(256, 128)
        self.up3 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.conv3 = DoubleConvV2(128, 64)
        self.outc = nn.Conv2d(64, 1, 1)

        self.ds1 = DSBlock(256)
        self.ds2 = DSBlock(128)

    def forward(self, image, v_map):
        x1 = self.inc(image)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)

        x4 = self.bottleneck_att(x4, v_map)

        d = self.up1(x4)
        d = self.conv1(torch.cat([d, self.att1(x3, v_map)], dim=1))
        ds1_out = self.ds1(d)

        d = self.up2(d)
        d = self.conv2(torch.cat([d, self.att2(x2, v_map)], dim=1))
        ds2_out = self.ds2(d)

        d = self.up3(d)
        d = self.conv3(torch.cat([d, self.att3(x1, v_map)], dim=1))
        main_out = self.outc(d)

        if self.training:
            return torch.sigmoid(main_out), torch.sigmoid(ds1_out), torch.sigmoid(ds2_out)
        return torch.sigmoid(main_out)

    def get_alphas(self):
        return {
            'bottleneck': torch.sigmoid(self.bottleneck_att.alpha).item(),
            'deep': torch.sigmoid(self.att1.alpha).item(),
            'mid':  torch.sigmoid(self.att2.alpha).item(),
            'fine': torch.sigmoid(self.att3.alpha).item(),
        }

class SoftSkeletonize(nn.Module):
    def __init__(self, num_iter=10):
        super().__init__()
        self.num_iter = num_iter
    def soft_erode(self, img): return -F.max_pool2d(-img, 3, stride=1, padding=1)
    def soft_dilate(self, img): return F.max_pool2d(img, 3, stride=1, padding=1)
    def soft_open(self, img): return self.soft_dilate(self.soft_erode(img))
    def forward(self, img):
        img_open = self.soft_open(img)
        skel = F.relu(img - img_open)
        curr_img = img
        for _ in range(self.num_iter):
            curr_img = self.soft_erode(curr_img)
            curr_open = self.soft_open(curr_img)
            delta = F.relu(curr_img - curr_open)
            skel = skel + F.relu(delta - skel * delta)
        return skel

class TrueSoftCLDiceLoss(nn.Module):
    def __init__(self, smooth=1.):
        super().__init__()
        self.smooth = smooth
        self.skel = SoftSkeletonize(num_iter=10)
    def forward(self, y_pred, y_true):
        skel_true = self.skel(y_true)
        skel_pred = self.skel(y_pred)
        tprec = (torch.sum(skel_pred * y_true) + self.smooth) / (torch.sum(skel_pred) + self.smooth)
        tsens = (torch.sum(skel_true * y_pred) + self.smooth) / (torch.sum(skel_true) + self.smooth)
        cl_dice = 2.0 * (tprec * tsens) / (tprec + tsens)
        return 1.0 - cl_dice

class FocalLoss(nn.Module):
    def __init__(self, alpha=0.75, gamma=2.0):
        super().__init__()
        self.alpha, self.gamma = alpha, gamma
    def forward(self, inputs, targets):
        bce = F.binary_cross_entropy(inputs, targets, reduction='none')
        p_t = inputs * targets + (1 - inputs) * (1 - targets)
        alpha_t = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        loss = alpha_t * (1 - p_t) ** self.gamma * bce
        return loss.mean()

class PCG_AdvancedLossV2(nn.Module):
    def __init__(self, cl_weight=0.3):
        super().__init__()
        self.cl_weight = cl_weight
        self.focal = FocalLoss()
        self.cldice = TrueSoftCLDiceLoss()
    def forward(self, pred, target):
        focal_loss = self.focal(pred, target)
        dice_loss = 1 - (2*(pred*target).sum() + 1e-8) / (pred.sum() + target.sum() + 1e-8)
        cld = self.cldice(pred, target)
        return (1 - self.cl_weight) * (focal_loss + dice_loss) + self.cl_weight * cld

# ==========================================
# [5] التدريب المتقدم (Advanced Training Loop)
# ==========================================
def train_model(model, train_loader, val_loader, device, model_name, epochs=80, patience=20):
    optimizer = torch.optim.AdamW(model.parameters(), lr=5e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer, max_lr=5e-4, epochs=epochs, steps_per_epoch=len(train_loader),
        pct_start=0.1, anneal_strategy='cos'
    )

    criterion = PCG_AdvancedLossV2(cl_weight=0.3)
    best_dice, counter, is_pcg = 0.0, 0, "PCG" in model_name
    save_path = f"{model_name}_BEST.pth"
    history = {'train_loss': [], 'val_dice': []}

    print(f"\n🚀 Training {model_name} | AdamW + OneCycleLR | Deep Sup: {'ON' if is_pcg else 'OFF'}")

    for epoch in range(epochs):
        model.train(); epoch_loss = 0
        ds_weight = max(0.1, 0.4 * (1 - epoch/epochs))

        for imgs, v_maps, masks in train_loader:
            imgs, v_maps, masks = imgs.to(device), v_maps.to(device), masks.to(device)
            optimizer.zero_grad()

            if is_pcg:
                main_out, ds1_out, ds2_out = model(imgs, v_maps)
                l_main = criterion(main_out, masks)
                l_ds1 = criterion(ds1_out, F.interpolate(masks, size=ds1_out.shape[2:], mode='nearest'))
                l_ds2 = criterion(ds2_out, F.interpolate(masks, size=ds2_out.shape[2:], mode='nearest'))
                loss = l_main + ds_weight * (l_ds1 + l_ds2)
            else:
                loss = criterion(model(imgs), masks)

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()
            epoch_loss += loss.item()

        avg_loss = epoch_loss / len(train_loader)
        history['train_loss'].append(avg_loss)

        # Validation
        model.eval(); val_dices = []
        with torch.no_grad():
            for imgs, v_maps, masks in val_loader:
                imgs, v_maps, masks = imgs.to(device), v_maps.to(device), masks.to(device)
                out = model(imgs, v_maps) if is_pcg else model(imgs)

                # 🟢 تعديل السكويز (Squeeze Fix)
                p = (out.squeeze(1).cpu().numpy() > 0.5).astype(bool)
                t = masks.squeeze(1).cpu().numpy().astype(bool)

                for j in range(p.shape[0]):
                    tp = np.logical_and(p[j], t[j]).sum()
                    fp = np.logical_and(p[j], ~t[j]).sum()
                    fn = np.logical_and(~p[j], t[j]).sum()
                    val_dices.append((2*tp) / (2*tp + fp + fn + 1e-8))

        avg_dice = np.mean(val_dices)
        history['val_dice'].append(avg_dice)

        marker = " ⭐ BEST" if avg_dice > best_dice else ""
        if marker: best_dice, counter = avg_dice, 0; torch.save(model.state_dict(), save_path)
        else: counter += 1

        print(f"  Ep [{epoch+1:2d}/{epochs}] Loss: {avg_loss:.4f} | Val Dice: {avg_dice:.4f} | LR: {optimizer.param_groups[0]['lr']:.6e} | DS_W: {ds_weight:.2f}{marker}")
        if counter >= patience: print("  ⏹️ Early Stop"); break

    return save_path, history

b_model = BaselineUNetV2().to(device)
path_b, h_base = train_model(b_model, train_loader, val_loader, device, "Baseline_UNetV2")

p_model = PCGNetV2().to(device)
path_p, h_pcg = train_model(p_model, train_loader, val_loader, device, "PCGNetV2")

# رسم منحنيات التدريب
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.plot(h_base['train_loss'], label='Baseline', linewidth=2)
ax1.plot(h_pcg['train_loss'], label='PCG-Net', linewidth=2)
ax1.set_title('Training Loss', fontsize=14, fontweight='bold')
ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss'); ax1.legend(); ax1.grid(True)

ax2.plot(h_base['val_dice'], label='Baseline', linewidth=2)
ax2.plot(h_pcg['val_dice'], label='PCG-Net', linewidth=2)
ax2.set_title('Validation Dice', fontsize=14, fontweight='bold')
ax2.set_xlabel('Epoch'); ax2.set_ylabel('Dice'); ax2.legend(); ax2.grid(True)
plt.tight_layout(); plt.show()

# ==========================================
# [6] التقييم العادل والمقارنة (Fair Evaluation)
# ==========================================
b_model.load_state_dict(torch.load('Baseline_UNetV2_BEST.pth', map_location=device))
b_model.eval()
p_model.load_state_dict(torch.load('PCGNetV2_BEST.pth', map_location=device))
p_model.eval()

def get_error_map(pred, target):
    err = np.zeros((pred.shape[0], pred.shape[1], 3), dtype=np.uint8)
    err[(pred==1) & (target==1)] = [255, 255, 255]
    err[(pred==1) & (target==0)] = [255, 0, 0]
    err[(pred==0) & (target==1)] = [0, 255, 255]
    return err

def remove_small_noise(mask_bool, min_size=50):
    mask_u8 = (mask_bool * 255).astype(np.uint8)
    num, labels, stats, _ = cv2.connectedComponentsWithStats(mask_u8, connectivity=8)
    clean = np.zeros_like(mask_u8)
    for i in range(1, num):
        if stats[i, cv2.CC_STAT_AREA] >= min_size: clean[labels == i] = 1
    return clean.astype(bool)

def visualize_fair(loader, b_mod, p_mod, dev, num=4):
    fig, axes = plt.subplots(num, 6, figsize=(24, 4 * num))
    fig.suptitle("Fair Comparison: Threshold=0.5 | Post-process: Remove areas < 50px", fontsize=16, fontweight='bold')
    cols = ['Original', 'Ground Truth', 'Baseline Pred', 'Baseline Error', 'PCG-Net Pred', 'PCG-Net Error']
    for ax, c in zip(axes[0], cols): ax.set_title(c, fontweight='bold')

    with torch.no_grad():
        for i, (imgs, vmaps, masks) in enumerate(loader):
            if i >= num: break
            img, vmap, mask = imgs[0:1].to(dev), vmaps[0:1].to(dev), masks[0:1].to(dev)

            # 🟢 تعديل السكويز (Squeeze Fix) لمنع انهيار الرسم
            p_base_raw = b_mod(img).squeeze(0).squeeze(0).cpu().numpy() > 0.5
            p_pcg_raw  = p_mod(img, vmap).squeeze(0).squeeze(0).cpu().numpy() > 0.5

            p_b_clean = remove_small_noise(p_base_raw, 50)
            p_p_clean = remove_small_noise(p_pcg_raw, 50)

            t_mask = mask.squeeze(0).squeeze(0).cpu().numpy().astype(bool)
            img_vis = img.squeeze(0).squeeze(0).cpu().numpy()

            err_b = get_error_map(p_b_clean, t_mask)
            err_p = get_error_map(p_p_clean, t_mask)

            axes[i,0].imshow(img_vis, cmap='gray'); axes[i,0].axis('off')
            axes[i,1].imshow(t_mask, cmap='gray'); axes[i,1].axis('off')
            axes[i,2].imshow(p_b_clean, cmap='gray'); axes[i,2].axis('off')
            axes[i,3].imshow(err_b); axes[i,3].axis('off')
            axes[i,4].imshow(p_p_clean, cmap='gray'); axes[i,4].axis('off')
            axes[i,5].imshow(err_p); axes[i,5].axis('off')
    plt.tight_layout()
    plt.show()

print("\n🖼️ جاري استخراج الصور العادلة...")
visualize_fair(test_loader, b_model, p_model, device, num=4)

# ==========================================
# [7] النتائج النهائية والإحصائيات (Final Stats)
# ==========================================
def compute_metrics(pred, true):
    tp = np.logical_and(pred, true).sum()
    fp = np.logical_and(pred, ~true).sum()
    fn = np.logical_and(~pred, true).sum()
    tn = np.logical_and(~pred, ~true).sum()
    return {
        'Dice': (2*tp) / (2*tp + fp + fn + 1e-8),
        'IoU': tp / (tp + fp + fn + 1e-8),
        'Sensitivity': tp / (tp + fn + 1e-8),
        'Specificity': tn / (tn + fp + 1e-8),
        'Precision': tp / (tp + fp + 1e-8),
        'Accuracy': (tp + tn) / (tp + tn + fp + fn + 1e-8)
    }

print("\n⏳ جاري حساب الإحصائيات لجميع صور الاختبار...")

base_metrics = {'Dice': [], 'IoU': [], 'Sensitivity': [], 'Specificity': [], 'Precision': [], 'Accuracy': []}
pcg_metrics = {'Dice': [], 'IoU': [], 'Sensitivity': [], 'Specificity': [], 'Precision': [], 'Accuracy': []}

with torch.no_grad():
    for imgs, vmaps, masks in tqdm(test_loader):
        imgs, vmaps, masks = imgs.to(device), vmaps.to(device), masks.to(device)

        pb_raw = b_model(imgs).squeeze(1).cpu().numpy() > 0.5
        pp_raw = p_model(imgs, vmaps).squeeze(1).cpu().numpy() > 0.5
        t_masks = masks.squeeze(1).cpu().numpy().astype(bool)

        for i in range(t_masks.shape[0]):
            t = t_masks[i]

            pb = remove_small_noise(pb_raw[i], 50)
            bm = compute_metrics(pb, t)
            for k in base_metrics: base_metrics[k].append(bm[k])

            pp = remove_small_noise(pp_raw[i], 50)
            pm = compute_metrics(pp, t)
            for k in pcg_metrics: pcg_metrics[k].append(pm[k])

# 🟢 إزالة تكرار الجداول كما طُلب والاكتفاء בחساب الإحصائيات
stat, p_val = stats.wilcoxon(base_metrics['Dice'], pcg_metrics['Dice'])

# ==========================================
# [8] جدول المقارنة الختامي ورسومات XAI (Final Results & Charts)
# ==========================================
improvement = np.mean(pcg_metrics['Dice']) - np.mean(base_metrics['Dice'])

results_data = {
    "Metric": ["Dice", "IoU", "Sensitivity", "Specificity", "Precision", "Accuracy"],
    "BaselineUNetV2": [
        f"{np.mean(base_metrics['Dice']):.4f}",
        f"{np.mean(base_metrics['IoU']):.4f}",
        f"{np.mean(base_metrics['Sensitivity']):.4f}",
        f"{np.mean(base_metrics['Specificity']):.4f}",
        f"{np.mean(base_metrics['Precision']):.4f}",
        f"{np.mean(base_metrics['Accuracy']):.4f}"
    ],
    "PCGNetV2 (Proposed)": [
        f"{np.mean(pcg_metrics['Dice']):.4f}",
        f"{np.mean(pcg_metrics['IoU']):.4f}",
        f"{np.mean(pcg_metrics['Sensitivity']):.4f}",
        f"{np.mean(pcg_metrics['Specificity']):.4f}",
        f"{np.mean(pcg_metrics['Precision']):.4f}",
        f"{np.mean(pcg_metrics['Accuracy']):.4f}"
    ],
    "Diff (Improvement)": [
        f"{np.mean(pcg_metrics['Dice']) - np.mean(base_metrics['Dice']):+.4f}",
        f"{np.mean(pcg_metrics['IoU']) - np.mean(base_metrics['IoU']):+.4f}",
        f"{np.mean(pcg_metrics['Sensitivity']) - np.mean(base_metrics['Sensitivity']):+.4f}",
        f"{np.mean(pcg_metrics['Specificity']) - np.mean(base_metrics['Specificity']):+.4f}",
        f"{np.mean(pcg_metrics['Precision']) - np.mean(base_metrics['Precision']):+.4f}",
        f"{np.mean(pcg_metrics['Accuracy']) - np.mean(base_metrics['Accuracy']):+.4f}"
    ]
}

df_results = pd.DataFrame(results_data)

print("\n\n" + "★"*85)
print(" "*25 + "🏆 التفاصيل الكاملة للمقاييس (All Metrics) 🏆")
print("★"*85)
print(df_results.to_string(index=False, justify='center'))
print("-" * 85)
print(f"🔬 P-Value (Wilcoxon Signed-Rank Test on Dice): {p_val:.4e}")

if p_val < 0.05 and improvement > 0:
    print("\n🌟 الخلاصة العلمية (Conclusion):")
    print("معمارية PCG-Net V2 أثبتت تفوقها الشامل على الـ Baseline في أغلب المقاييس الهامة.")
    print("التحسن في مقياس Dice يعتبر ذو دلالة إحصائية قوية (Statistically Significant)،")
    print("مما يؤكد فعالية المكونات المقترحة (Deep Supervision, Spatial-Channel Att, TrueSoftCLDice).")
else:
    print("\n📊 الخلاصة: النتائج الإجمالية متقاربة. يمكن مراجعة المنحنيات (Learning Curves) لمعرفة ما إذا كان النموذج يحتاج المزيد من التدريب (Epochs).")
print("★"*85)

alphas = p_model.get_alphas()
print("\n🧠 XAI Attention Weights (Sigmoid Alpha):")
for k, v in alphas.items(): print(f" - {k.capitalize()}: {v:.4f}")

# 🟢 إضافة الرسم البياني المفقود لـ XAI
plt.figure(figsize=(8, 5))
colors = ['#e74c3c', '#ff9999', '#66b3ff', '#99ff99']
bars = plt.bar(list(alphas.keys()), list(alphas.values()), color=colors, edgecolor='black')
plt.ylim(0, 1.0)
plt.title("XAI: Learned Attention Weights per Level", fontsize=14, fontweight='bold')
plt.ylabel('Sigmoid(α)', fontsize=12)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
             f'{bar.get_height():.4f}', ha='center', va='bottom', fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
